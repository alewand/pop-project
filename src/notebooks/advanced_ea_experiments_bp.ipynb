### Funckje pomocnicze

```python
import numpy as np
import pandas as pd
from pandera.typing import DataFrame

def ea_run_multiple_runs_df(
    pokemons: DataFrame[PokemonSchema],
    solver: EvolutionaryAlgorithmPokemonSolver,
    opponents: list[PokemonTeam] | None = None,
    runs: int = 8,
) -> tuple[pd.DataFrame, pd.DataFrame]:
    """
    Returns:
      df_runs: one row per run with best team info
      df_summary: aggregated stats (mean/median/std/min/max/count)
    """
    rows = []

    for i in range(runs):
        print(f"EA Run {i+1}/{runs}")
        best_team, best_fitness, _, _ = solver.solve(pokemons, opponents=opponents)

        print(f"Best fitness: {best_fitness}")
        print(f"Best team: {best_team}")

        # WAŻNE ŻEBY REPREZENTACJA DRUŻYNY BYŁA POD NAZWĄ 'pokemons', bo tam df jest pod spodem resizowany żeby w raporcie się zmieściło !!!!! !!!!!
        rows.append(
            {
                "run": i,
                "solver": "EA",
                "elite_size": solver.elite_size,
                "fitness": float(best_fitness),
                "stats_sum": int(best_team.get_stats_sum()),
                "pokemons": repr(best_team),
            }
        )

    df_runs = pd.DataFrame(rows)

    df_summary = (
        df_runs.groupby("solver")["fitness"]
        .agg(["mean", "median", "std", "min", "max", "count"])
        .reset_index()
        .sort_values("mean", ascending=False)
        .round(6)
    )

    return df_runs, df_summary
```

### Opponents

```python
opponents = PokemonTeam.generate_unique_teams(
                pokemons,
                oponents_limit=100,
                max_attempts=300,
                team_size=6,
                unique_types=True
            )
```

#### EA with different mutation rate

```python
MUTATION_RATES = [0.1, 0.3, 0.5, 0.7, 0.9]


for mutation_rate in MUTATION_RATES:
    solver = EvolutionaryAlgorithmPokemonSolver(
        mutation_rate=mutation_rate
    )

    opponents_typings = visualize_opponents_typing_distribution(opponents)
    opponents_stats = visualize_opponents_stat_sums_violin(opponents)

    ea_runs_df, ea_summary_df = ea_run_multiple_runs_df(
        pokemons=pokemons,
        solver=solver,
        runs=6,
    )

    report = report = PdfReport(str(REPORT_DIR / f"ea_experiment_mutation_rate_{mutation_rate}.pdf"), title="Evolutionary Algorithm - Mutation Rate Experiment")

    report.add_text(
        f"""Setup:
         - Population size: {solver.population_size}
         - Generations: {solver.generations}
         - Mutation rate: {solver.mutation_rate}
         - Elite size: {solver.elite_size}
         - Tournament size: {solver.tournament_size}
    """
    )

    report.add_figure(opponents_typings)
    report.add_figure(opponents_stats)

    report.add_dataframe(ea_runs_df, title="EA - Results of individual runs")
    report.add_dataframe(ea_summary_df, title="EA - Summary statistics")

    report.write()
```

### EA with different population sizes

```python
POPULATION_SIZES = [20, 50, 100, 200]

for population_size in POPULATION_SIZES:
    solver = EvolutionaryAlgorithmPokemonSolver(
        population_size=population_size
    )

    opponents_typings = visualize_opponents_typing_distribution(opponents)
    opponents_stats = visualize_opponents_stat_sums_violin(opponents)

    ea_runs_df, ea_summary_df = ea_run_multiple_runs_df(
        pokemons=pokemons,
        solver=solver,
        runs=6,
    )
    report = report = PdfReport(str(REPORT_DIR / f"ea_experiment_population_size_{population_size}.pdf"), title="Evolutionary Algorithm - Population Size Experiment")
    report.add_text(
        f"""Setup:
         - Population size: {solver.population_size}
         - Generations: {solver.generations}
         - Mutation rate: {solver.mutation_rate}
         - Elite size: {solver.elite_size}
         - Tournament size: {solver.tournament_size}
    """
    )

    report.add_figure(opponents_typings)
    report.add_figure(opponents_stats)

    report.add_dataframe(ea_runs_df, title="EA - Results of individual runs")
    report.add_dataframe(ea_summary_df, title="EA - Summary statistics")
    report.write()
```
