### Inicjalizacjia pokemonów

```python
pokemons = get_pokemons()  # Ostatnie ewolucje

print("Liczba pokemonów:", len(pokemons))
print("Kolumny:", list(pokemons.columns)[:20], "...")
display(pokemons.head(3))
```

### Tworzenie drużyn na podstawie nazw

```python
def create_pokemon_tean_based_on_names(names):
    members  = pokemons[pokemons['name'].isin(names)]
    return PokemonTeam(members)
```

### Walka

```python
def compare_teams(team1, team2):
    result = simulate_battle(team1, team2,
                             multiply_type_multiplier,
                             damage_attack_devide_defense,)

    print("Team1:", team1)
    print("Team2:", team2)
    print("Team1 remaining HP:", result, " / ", sum(team1.get_hps()))

    print("Winner is Team1" if result > 0 else "Winner is Team2")

    return team1 if result > 0 else team2
```

### Inicjalizacja drużyn

```python
best_ea_team_names = ['Snorlax', 'Salamence', 'Gigalith', 'Wobbuffet', 'Lapras', 'Galvantula']
best_sa_team_names = ['Gogoat', 'Rhyperior', 'Ampharos', 'Blissey', 'Drifblim', 'Wailord']
```

```python
sa_team = create_pokemon_tean_based_on_names(best_sa_team_names)
eq_team = create_pokemon_tean_based_on_names(best_eq_team_names)

print("SA team:", sa_team)
print("EA team:", eq_team)
```

### Symulacja walki

```python
winner = compare_teams(sa_team, eq_team)
```
